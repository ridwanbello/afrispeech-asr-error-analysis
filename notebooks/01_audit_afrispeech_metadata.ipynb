```python
from pathlib import Path

PROJECT_ROOT = Path.cwd()

RESULTS_DIR = PROJECT_ROOT / "results"
METRICS_DIR = RESULTS_DIR / "metrics"
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

METRICS_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# Load Afrispeech data

import pandas as pd
from datasets import load_from_disk, Audio

ds = load_from_disk("../data/raw/afrispeech_arrow_16k")
print("Available splits:", ds.keys())
```

```python
# Load test data
# Load test data

test_ds = ds["test"]

print(test_ds)
print("Number of test samples:", len(test_ds))
print("Columns:", test_ds.column_names)
```

```python
# Convert test metadata to DataFrame

test_df = test_ds.remove_columns(["audio"]).to_pandas()

print("Test metadata shape:", test_df.shape)
display(test_df.head())
```

```python
# Load clinical test data
# Load clinical test data

clinical_df = test_df[test_df["domain"].str.lower() == "clinical"].copy()

print("Clinical test samples:", len(clinical_df))
display(clinical_df.head())
```

```python
# Load general test data
general_df = test_df[test_df["domain"].str.lower() == "general"].copy()

print("General test samples:", len(general_df))
display(general_df.head())
```

```python
# View column details

column_summary = pd.DataFrame({
    "column": test_df.columns,
    "dtype": [test_df[col].dtype for col in test_df.columns],
    "missing_count": [test_df[col].isna().sum() for col in test_df.columns],
    "missing_percent": [round(test_df[col].isna().mean() * 100, 2) for col in test_df.columns],
    "unique_values": [test_df[col].nunique(dropna=True) for col in test_df.columns],
})

display(column_summary)

column_summary.to_csv(METRICS_DIR / "metadata_column_summary.csv", index=False)
```

```python
# Sample counts per accent for clinical domain
# Sample counts by domain

domain_counts = (
    test_df["domain"]
    .value_counts(dropna=False)
    .reset_index()
)

domain_counts.columns = ["domain", "sample_count"]

display(domain_counts)

domain_counts.to_csv(METRICS_DIR / "sample_counts_by_domain.csv", index=False)
```

```python
# Sample counts per accent for clinical domain

clinical_accent_counts = (
    clinical_df["accent"]
    .value_counts(dropna=False)
    .reset_index()
)

clinical_accent_counts.columns = ["accent", "clinical_sample_count"]

display(clinical_accent_counts.head(30))

clinical_accent_counts.to_csv(
    METRICS_DIR / "clinical_sample_counts_by_accent.csv",
    index=False
)
```

```python
# Check any missing data on transcripts
# Sample counts per accent for general domain

general_accent_counts = (
    general_df["accent"]
    .value_counts(dropna=False)
    .reset_index()
)

general_accent_counts.columns = ["accent", "general_sample_count"]

display(general_accent_counts.head(30))

general_accent_counts.to_csv(
    METRICS_DIR / "general_sample_counts_by_accent.csv",
    index=False
)
```

```python
# Find duplicated transcript texts
duplicate_transcripts = (
    test_df[test_df["transcript"].duplicated(keep=False)]
    .sort_values("transcript")
)

print("Rows with duplicated transcript text:", len(duplicate_transcripts))
print("Number of repeated transcript texts:", duplicate_transcripts["transcript"].nunique())

display(
    duplicate_transcripts[
        ["audio_id", "speaker_id", "domain", "accent", "country", "duration", "transcript"]
    ].head(30)
)
```

```python
from IPython.display import display, Audio, Markdown

def play_audio_by_index(idx):
    sample = test_ds[int(idx)]
    audio = sample["audio"]

    print("Index:", idx)
    print("Audio ID:", sample["audio_id"])
    print("Speaker ID:", sample["speaker_id"])
    print("Domain:", sample["domain"])
    print("Accent:", sample["accent"])
    print("Country:", sample["country"])
    print("Duration:", sample["duration"])
    print("Transcript:", sample["transcript"])

    display(Audio(audio["array"], rate=audio["sampling_rate"]))
```

```python
play_audio_by_index(4290)
play_audio_by_index(4866)
```

```python
# Minimum and maximum duration by domain

duration_summary = (
    clinical_df
    .groupby("domain")
    .agg(
        samples=("audio_id", "count"),
        min_duration_seconds=("duration", "min"),
        max_duration_seconds=("duration", "max"),
        mean_duration_seconds=("duration", "mean"),
        median_duration_seconds=("duration", "median"),
        total_duration_hours=("duration", lambda x: x.sum() / 3600),
    )
    .reset_index()
)

# Round numeric columns
duration_summary[
    [
        "min_duration_seconds",
        "max_duration_seconds",
        "mean_duration_seconds",
        "median_duration_seconds",
        "total_duration_hours",
    ]
] = duration_summary[
    [
        "min_duration_seconds",
        "max_duration_seconds",
        "mean_duration_seconds",
        "median_duration_seconds",
        "total_duration_hours",
    ]
].round(2)

display(duration_summary)
```
