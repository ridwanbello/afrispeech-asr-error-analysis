```python
from pathlib import Path
from datasets import load_from_disk
import shutil

PROJECT_ROOT = Path("/home/ridwan/projects/afrispeech-project")

DATASET_PATH = PROJECT_ROOT / "data" / "raw" / "afrispeech_arrow_16k"
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

print("Project root:", PROJECT_ROOT)
print("Dataset path:", DATASET_PATH)
print("Dataset exists:", DATASET_PATH.exists())
```

```python
ds = load_from_disk(str(DATASET_PATH))
test_ds = ds["test"]

print(test_ds)
print(test_ds.column_names)
```

```python
clinical_test_ds = test_ds.filter(
    lambda x: str(x["domain"]).lower().strip() == "clinical"
)

general_test_ds = test_ds.filter(
    lambda x: str(x["domain"]).lower().strip() == "general"
)

print("Clinical samples:", len(clinical_test_ds))
print("General samples:", len(general_test_ds))
print("Combined:", len(clinical_test_ds) + len(general_test_ds))
print("Original test samples:", len(test_ds))
```

```python
clinical_arrow_path = PROCESSED_DIR / "clinical_test_arrow"
general_arrow_path = PROCESSED_DIR / "general_test_arrow"

for path in [clinical_arrow_path, general_arrow_path]:
    if path.exists():
        shutil.rmtree(path)
        print("Removed old version:", path)

clinical_test_ds.save_to_disk(str(clinical_arrow_path))
general_test_ds.save_to_disk(str(general_arrow_path))

print("Saved clinical Arrow split:", clinical_arrow_path)
print("Saved general Arrow split:", general_arrow_path)
```

```python
clinical_check = load_from_disk(str(clinical_arrow_path))
general_check = load_from_disk(str(general_arrow_path))

print("Reloaded clinical:", clinical_check)
print("Reloaded general:", general_check)

print("Clinical samples:", len(clinical_check))
print("General samples:", len(general_check))
```
